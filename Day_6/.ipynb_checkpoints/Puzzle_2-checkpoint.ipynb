```python
f=open("input")
orbits=f.read()
f.close()
orbits = orbits.split("\n")
orbits = [o.split(')') for o in orbits]
orbits = [o for o in orbits if o!=['']]
```

```python
def transfers_count(current_node, target_node, seen_nodes, orbits):
    queued_nodes = []
    for o in orbits:
        if o[0]==current_node or o[1]==current_node:
            if o[0]==target_node or o[1]==target_node:
                return 1
            if o[1] not in seen_nodes:
                queued_nodes.append(o[1]) 
                seen_nodes.append(o[1]) 
            if o[0] not in seen_nodes:
                queued_nodes.append(o[0])
                seen_nodes.append(o[0])
    while True:
        try:
            next_node = queued_nodes.pop(0)
            transfers = transfers_count(next_node, target_node, seen_nodes, orbits)
            if transfers>0:
                return transfers+1
        except:
            return 0
transfers_count("YOU", "SAN", ["YOU"],orbits)-2
```

```python
def transfers_count(current_node, target_node, seen_nodes, orbits):
    queued_nodes = []
    for o in orbits:
        if o[0]==current_node:
            if o[1]==target_node:
                return 1
            if o[1] not in seen_nodes:
                queued_nodes.append(o[1]) 
                seen_nodes.append(o[1]) 
        if o[1]==current_node:
            if o[0]==target_node:
                return 1
            if o[0] not in seen_nodes:
                queued_nodes.append(o[0])
                seen_nodes.append(o[0])
    while True:
        try:
            next_node = queued_nodes.pop(0)
            transfers = transfers_count(next_node, target_node, seen_nodes, orbits)
            if transfers>0:
                return transfers+1
        except:
            return 0
transfers_count("YOU", "SAN", ["YOU"],orbits)-2
```

```python
def orbit_count(current_node, depth, orbits):
    queued_nodes = []
    count=0
    for o in orbits:
        if o[0] == current_node:
            queued_nodes.append(o[1])
            count+=1+depth
    while True:
        try:
            next_node = queued_nodes.pop(0)
            count+=orbit_count(next_node, depth+1, orbits)
        except:
            break
    return count
orbit_count("COM", 0, orbits)
```
