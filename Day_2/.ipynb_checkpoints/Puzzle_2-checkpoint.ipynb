```python
f = open("input")
intcode = [int(i) for i in f.read().split(",")]
f.close()
#All possible pairs of inputs:
options = list(itertools.product(range(0,len(intcode)), range(0,len(intcode))))
for option in options:
    f = open("input")
    intcode = [int(i) for i in f.read().split(",")]
    f.close()
    program_counter=0
    intcode[1]=option[0]
    intcode[2]=option[1]
    #Execute the program
    while True:
        opcode=intcode[program_counter]
        param_1=intcode[program_counter+1]
        param_2=intcode[program_counter+2]
        param_3=intcode[program_counter+3]
        #Extremely unefficient switch:
        if opcode==1:
            intcode[param_3]=intcode[param_1]+intcode[param_2]
        elif opcode==2:
            intcode[param_3]=intcode[param_1]*intcode[param_2]
        elif opcode==99:
            print("Program complete")
            break
        else:
            print("An error ha occurred")
            break
        program_counter+=4
    if intcode[0]==19690720:
        print("Found correct input: " + str(option))
        print("Solution is: " + str(100*option[0]+option[1]))
        break;
        
print(intcode)
```
